# Tunneling Jupyter Notebooks and MLFlow on Jetstream2

* You can access mlflow and jupyter notebooks on Jetstream2

## Jupyter

To use a jupyter notebook on Jetstream2, you'll want to have Jupyterlab installed on a running instance. I have it installed in a Conda environment:

```
name: jetstream-jupyter
channels:
  - conda-forge
  - defaults

dependencies:
  - python=3.11
  - jupyterlab
  - ipykernel
  - numpy
  - pandas
  - matplotlib
  - scipy
  - xarray
  - netCDF4
  - dask
  - h5py
  - pip
  - pip:
      - jupyterlab-git
```

Next, login to your Jetstream2 instance and start up a JupyterLab server on port 8888.
```
jupyter lab --no-browser --ip=127.0.0.1 --port=8888
```
Once it's running, you should see some output like:
```
   To access the server, open this file in a browser:
        file:///home/exouser/.local/share/jupyter/runtime/jpserver-195725-open.html
    Or copy and paste one of these URLs:
        http://127.0.0.1:8888/lab?token=dc62b9d2081a1785e2a3a82d89585ea117f5f145eb48726b
```

On your local machine, ssh into Jetstream2, enabling port forwarding.
```
ssh -L 8888:127.0.0.1:8888 youruser@your.jetstream2.ip.address
```



You can copy the output from the JupyterLab server (e.g.  http://127.0.0.1:8888/lab?token=dc62b9d2081a1785e2a3a82d89585ea117f5f145eb48726b) into your local browser. You should now be able to access JupyterLab.

## MLFlow
The process for tunneling MLFlow is similar. On your JetStream2 instance, creat an environment with MLFlow installed. Then launch an MLFlow server on the instance.
```
mlflow ui --host 127.0.0.1 --port 5000 --backend-store-uri ./mlruns
```
In my case, I have my MLFlow output in a folder `.mlruns` in the directory where I'm launching the mlflow server.

On your local machine, you'll need to tunnel port 5000 to access MLFlow.
```
ssh -L 5000:127.0.0.1:5000 youruser@your.jetstream2.ip.address
```
You may also tunnel multiple ports at one for Jupyter and MLFlow.
```
ssh -L 5000:127.0.0.1:5000 -L 8888:127.0.0.1:8888 youruser@your.jetstream2.ip.address
```

Now, you should be ble to access MLFlow in your browser from the address
```
127.0.0.1:8888
```
